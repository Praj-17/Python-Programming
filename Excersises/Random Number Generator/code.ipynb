## Generating Random numbers


## Documentation -


##### Methodology :
 - Here I have used the python inbuilt Id() function to genrate random numbers
 - An object is always created with a unique id
 - this unique id is is in the form of an 13 digit integer
 - Now, once we get this 13 digit integer we can manipulate it using any of the arithmentic methods such as Modulus, division, square, cube, etc.
 - To create a demo function I have taken the square of the number and then sliced that square to take it's middle digits
 - This is just a random technique there might be some other techniques like directly taking the middle term of the 13 digit id or dividing it by a certain by number and then taking its middle digits or by any other suitable arithmetic method.

##### Demo Function Algorithm
- use the id() function to get the 13 digit id
- square the id and get it's middle terms
-  the square will have around 26 digits and since we dont want such huge number we will slice it to smaller numbers for example i have taken 2 middle digits to create the numbers
- sometimes it returns 3 digit number this randomness is created due to having more bigger digits such as 7,8,9 in the id causing it to have more than 26 digits.
- at the end I have run a for loop to specify the number of integers that I have been wanting here as an example I have taken 100 numbers
- You can increase this for loop count to generate some more numbers.

```python
def demo_function(num):
        num = id(num)
        num = int(str(num*num)[11:-11])
        return num
```

```python
nums = []
for i in range(1 ,100):
    print(demo_function(i))
    nums.append(demo_function(i))
    
    
```

# Predicting the value of PI


###### The idea


Imagine throwing darts at a darts board. If you know that you will hit the board every single time but the location of the dart is uniformly distributed, then how what is the likelihood of the dart ending up in the top half of the board? Intuitively, it’s 50%. Why? Because the area of the top half is half of the bottom area.

We’ll exploit this fact and start drawing 2 random numbers at a time (also called a point in 2D or a vector) within a square of (-1,-1) and (1, 1), like this one:




https://miro.medium.com/max/750/1*hnS7GrHH-ucp5qR47_E9nA.png

Conveniently, this square fits a unit circle with radius = 1. So if the length of the vector is bigger than 1 then our point is outside the circle, if it’s less than 1 then it must be inside the unit circle.

https://miro.medium.com/max/750/1*n8FEbFTL8c2xu3E2trU2lw.png

We then draw random points many times and count how many times we draw a point from inside the circle vs outside the circle. That ratio should be close to the ratio of the areas of the circle and the square:

circle = pi*r*r
square = 4*r*r

circle/square = pi/4

pi = 4 * circle/square

Circle and Square represent the area of the shapes.
That is, we get pi by taking the number of points inside the circle divided by the number of points inside the square and multiplying by 4.




```python
import math
def approximate_pi():
    total_points = 0
    within_circle = 0
    for i in range (10000):
        x = demo_function(i)
        y = demo_function(i)
        total_points += 1
        distance = math.sqrt(x**2+y**2)
        if distance < 1050:
            
            within_circle += 1
        if total_points % 1000 == 0:
            pi_estimate = 4 * within_circle / total_points
            yield pi_estimate

```
